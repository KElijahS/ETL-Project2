```python
import pandas as pd
```

```python
# make paths and read csv raw data
data_path = "Dataset/pokedex_(Update_04.21)_Unclean.csv"

data = pd.read_csv(data_path)
```

```python
#there are several columns with tons of nul values so I got rid of them
#a couple were only missing under 5 so i kept those in (weight_kg, ability_1, growth_rate, egg_type_1, egg_cycle)
#get needed columns
pokedex_df = data[['pokedex_number','name','german_name','japanese_name','generation','status','type_1','height_m','weight_kg','ability_1','total_points','hp','attack','defense','sp_attack','sp_defense','speed','growth_rate','egg_type_1','egg_cycles','against_normal','against_fire','against_water','against_electric','against_grass','against_ice','against_fight','against_poison','against_ground','against_flying','against_psychic','against_bug','against_rock','against_ghost','against_dragon','against_dark','against_steel','against_fairy']]

pokedex_df.count()
```

```python
#drop all null values
pokedex_df = pokedex_df.dropna()

pokedex_df.count()
```
